**Döngüler**

* range
* for
* while 

range(start_value, end_value, step)

Belirtilmediği durumlarda;
**start_value** = 0,
**step** = 1 kabul edilir.
end_value dahil değildir.

```python
range(5) # range(0, 5, 1)

range(2, 34, 3)

range(100, 1, -5)
```

```python
for item in range(5):
  print(item)
```

```python
for i in range(0, 34, 3):
  print(i)
```

```python
for x in range(40, 1, -5):
  print(x)
```

```python
#break - istenen koşul sağlandığında döngüyü kırar ve çıkar
for x in range(40, 1, -5):
  if x <= 20:
    break
  print(x)

print("Bitti")
```

```python
#continue - istenen koşulu atlayıp döngüye devam eder
for x in range(40, 1, -5):
  if x % 2 == 0:
    continue
  print(x)
```

**while** Koşul sağlandığı sürece çalışan döngü

```python
count = 1
while count < 5:
  print(count)
  count += 1
  
print("Döngü tamamlandı")
```

```python
#sonsuz döngü
count = 1
while count < 5:
  print(count)
```

```python
number = 5
for i in range(number):
  for j in range(i + 1):
    print("*", end=" ") # end ifadesi yan yana yazmaya yarar
  print()
```

```python
# Kullanıcıdan bir sayı alıp ardından 0 ile girilen değer
# arasındaki sayıların toplamını veren kod blogunu yazınız

number = int(input("Bir sayı giriniz: "))
total = 0
for i in range(1, number + 1):
  total += i

print("0 ile {} sayıları arasındaki sayıların toplamı {}".format(number, total))
```

```python
# Toplama işlemi esnasında eğer o anki değer 
# 5 e tam bölünebiliyorsa o sayıyı toplama ekleme

number = int(input("Bir sayı giriniz: "))
total = 0

for item in range(1, number + 1):
  if item % 5 == 0:
    continue
  total += item

print(f"0 ile {number} sayıları arasındaki sayıların toplamı {total}")
```

```python
# Bir sayının kaç basamaklı olduğunu hesaplayalım

x = 6200
count = 0

while x != 0:
  count += 1
  x = x // 10

print(f"Sayı {count} basamaklıdır.")
```

```python
# Bir aralık belirleyip bu aralıktaki tüm asal sayıları ekrana yazın
start_value = 3
end_value = 20

for i in range(start_value, end_value):
  asal = True
  for j in range(2, i):
    if i % j == 0:
      asal = False
      break
  if asal:
    print(i)
```

```python
#Fibonacci serisini ekrana bastıran kod blogunu
num1, num2 = 0, 1
x = int(input("Seri için bir sayı giriniz: "))

for i in range(x):
  print(num1, end=" ")
  total = num1 + num2
  num1 = num2
  num2 = total
```

```python
# faktöriyel hesabı 
number = int(input("Bir sayı giriniz: "))
x = 1
for i in range(number, 1, -1):
  x *= i

print(f"{number} sayısının faktöriyeli = {x}")
```

```python
#Girilen sayıyı tersine çevirelim
a = 5689
print("Girilen sayı: ", a)
new_number = 0
while a > 0:
  temp = a % 10
  new_number = new_number * 10 + temp
  a //= 10

print("Ters çevrilmiş hali: ", new_number)
```
